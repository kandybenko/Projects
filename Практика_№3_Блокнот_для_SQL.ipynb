```python
# Обозначим в переменную путь к БД
path_db = "/content/pizza_orders.db"
```

```python
import sqlite3
from pathlib import Path
import pandas as pd

# путь к файлу БД
db_path = Path(path_db)  # ваш файл

# 1. подключение (создаст файл, если его нет)
con = sqlite3.connect(db_path)

# 2. полезно: включить внешние ключи и доступ к колонкам по имени
con.execute("PRAGMA foreign_keys = ON;")
con.row_factory = sqlite3.Row  # строки как dict-похожие объекты
```

# **Соединение с БД в Python**

```python
query = """
    SELECT *
    FROM orders;
"""

pd.read_sql(query, con)
```

```python
query = """
    SELECT
        DATE(order_datetime) as date,
        TIME(order_datetime) as time
    FROM orders;
"""

pd.read_sql(query, con)
```
